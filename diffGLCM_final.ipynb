```python
import torch
from torch import sigmoid
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import graycomatrix


min_val = 0
max_val = 9
bin_num = 8
interval_length = (max_val - min_val) / bin_num
kernel_width = interval_length / (40)

def calc_activation_maps(img):
    # apply approximated shifted rect (bin_num) functions on img
    
    bins_min_max = np.linspace(min_val, max_val, bin_num + 1)
    bins_av = (bins_min_max[0:-1] + bins_min_max[1:]) / 2
    bins_av = torch.tensor(bins_av, dtype = torch.float32)  # shape = (,bin_num)
    bins_av = torch.unsqueeze(bins_av, 0)  # shape = (1,bin_num)
    bins_av = torch.unsqueeze(bins_av, 0)  # shape = (1,1,bin_num)
    
    # Currently passing pre-flattened images
    m = nn.Flatten()
    # img_flat = torch.unsqueeze(m(img), -1)
    
    # img_flat = torch.unsqueeze(img, -1)

    maps = activation_func(img_flat, bins_av)  # shape = (batch_size,H*W,bin_num)
    return maps

def activation_func(img_flat, bins_av):
    # torch.set_default_device(self.device)

    img_minus_bins_av = torch.sub(img_flat, bins_av)  # shape=  (batch_size,H*W,bin_num)
    img_plus_bins_av = torch.add(img_flat, bins_av)  # shape = (batch_size,H*W,bin_num)
    maps = sigmoid((img_minus_bins_av + interval_length / 2) / kernel_width) \
            - sigmoid((img_minus_bins_av - interval_length / 2) / kernel_width) \
            + sigmoid((img_plus_bins_av - 2 * min_val + interval_length / 2) / kernel_width) \
            - sigmoid((img_plus_bins_av - 2 * min_val - interval_length / 2) / kernel_width) \
            + sigmoid((img_plus_bins_av - 2 * max_val + interval_length / 2) / kernel_width) \
            - sigmoid((img_plus_bins_av - 2 * max_val - interval_length / 2) / kernel_width)
    return maps
```

```python
#Row wise calculation

mat_original = torch.tensor([[1,1,5,6,8], [2,3,5,7,1], [4,5,7,1,2], [8,5,1,2,5]], dtype=torch.float32)

mat_no_shift = mat_original[:,:-1]
mat_shift = mat_original[:,1:]

mat_no_shift_flatten = torch.reshape(mat_no_shift, (-1,)) 
mat_shift_flatten = torch.reshape(mat_shift, (-1,)) 

# mat_shift = torch.tensor([1,5,6,8 ,3,5,7,1 ,5,7,1,2 ,5,1,2,5], dtype=torch.float32)
print(mat_no_shift_flatten)
print(mat_shift_flatten)
```

```python
#For Batch Simulation
mat_no_shift_flatten = torch.unsqueeze(mat_no_shift_flatten, 0)
mat_shift_flatten = torch.unsqueeze(mat_shift_flatten, 0)

print(mat_no_shift_flatten.shape, mat_shift_flatten.shape)
hist_1 = calc_activation_maps(mat_no_shift_flatten)
hist_2 = calc_activation_maps(mat_shift_flatten)
print(hist_1.shape, hist_2.shape)
```
